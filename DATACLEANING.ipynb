# Cleaning data

## Empty cells

#### Remove Rows
One way to deal with empty cells is to remove rows that contain empty cells.

This is usually OK, since data sets can be very big, and removing a few rows will not have a big impact on the result.

```python
import pandas as pd
ufo = pd.read_table('http://bit.ly/uforeports',sep = ',')

ufo.info()
```

```python
new_ufo=ufo.dropna() # By default, the dropna() method returns a new DataFrame, and will not change the original.
```

```python
new_ufo.head()
```

```python
new_ufo.info() # look now , the rows with raw values is removed
```

#### Replace empty values

```python
replaced_ufo = ufo.fillna({'City':'city not reported'})
replaced_ufo.head()
```

### Removing Duplicates

```python
import pandas as pd
data = pd.DataFrame(

    {
        'Name':['Hasmukh','Arvind','Vijay','Shailesh','Shailesh','Arvind'],
        'Sirname':['Kheni','Lathiya','Kaklotar','Makwana','Makwana','Makwana'],
        'Age':['40','44','34','24','24','55']
        
    }
)

without_duplicates = data.drop_duplicates()
print(without_duplicates)
```
