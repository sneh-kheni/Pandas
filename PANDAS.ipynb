```python
#series
a = [1,2,3,4]
b = pd.Series(a)
print(b[0])
print(b)

#Creating labesls
b = pd.Series(a,index=['x','y','z','a'])
print(b)
#When you have created labels, you can access an item by referring to the label.
print(b['a'])

#Key/Value Objects as Series
#You can also use a key/value object, like a dictionary, when creating a Series.
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)
#To select only some of the items in the dictionary, use the index argument and specify only the items you want to include in the Series.
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories,index=["day1","day2"])
print(myvar)
```

```python
# Difference between .loc and .iloc (simple explanation)

# .loc → label-based (use names)
# .iloc → index-based (use numbers)

# 🔹 Example 1: .loc (uses row/column names)
print("Using .loc:")
print(ufo.loc[0:2, ['City', 'State']])   # rows with index labels 0,1,2 → columns City & State
# 👉 .loc includes the last label (2)

# 🔹 Example 2: .iloc (uses row/column positions)
print("\nUsing .iloc:")
print(ufo.iloc[0:2, [0, 1]])             # rows 0 & 1 → first two columns
# 👉 .iloc excludes the last index (2)
```

```python
import pandas as pd
import numpy as np

data = {
    'City': ['New York', 'los angeles ', 'Chicago', np.nan, 'Houston', 'Phoenix', '   Dallas'],
    'State': ['NY', 'CA', 'IL', 'TX', np.nan, 'AZ', 'TX'],
    'Shape Reported': ['DISK', 'light', np.nan, 'CIRCLE', 'DISK', 'disk', 'CIRCLE'],
    'Colors Reported': ['RED', np.nan, 'BLUE', 'green', 'RED', 'RED', np.nan],
    'Time': ['6/1/1930 22:00', '6/30/1930 20:00', np.nan, '7/2/1930 19:00', '7/3/1930 21:00', 'invalid_date', '7/4/1930 20:00']
}

ufo = pd.DataFrame(data)
```

```python
#1
ufo.isnull().sum()
#2
ufo.fillna({'Colors Reported':'Unknown'})
#3
bool_series=pd.notnull(ufo['Time'])
ufo = ufo[bool_series]
#4
ufo['City'] = ufo['City'].str.strip()
#5
ufo['City'] = ufo['City'].str.capitalize()
#7
ufo.rename(columns={'Shape Reported':'SHAPE REPORTED',"Colors Reported":"COLORS REPORTED"},inplace=False)
#8
#not any duplicates
reset_indexes = ufo.reset_index()
reset_indexes
```
